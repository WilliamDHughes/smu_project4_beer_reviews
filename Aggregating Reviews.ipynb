```python
import pandas as pd
```

```python
df = pd.read_csv("Resources/Clean/reviews_merged.csv")
```

```python
df.head()
```

```python
df.availability.value_counts()
```

```python
df = df.drop(df[df['availability'] == 'Limited (brewed once)'].index)
```

```python
df.info()
```

```python
df.beer_id.nunique()
```

```python
agg = pd.DataFrame()

for beer in df['beer_id'].unique():
    reviews = df.loc[df.beer_id == beer]
    beer_id = reviews.iloc[0].beer_id
    beer_name = reviews.iloc[0].beer_name
    beer_style = reviews.iloc[0].beer_style
    look = reviews.look.mean()
    smell = reviews.smell.mean()
    taste = reviews.taste.mean()
    feel = reviews.feel.mean()
    overall = reviews.overall.mean()
    availability = reviews.iloc[0].availability
    abv = reviews.iloc[0].abv
    brewery_id = reviews.iloc[0].brewery_id
    brewery_name = reviews.iloc[0].brewery_name
    city = reviews.iloc[0].city
    state = reviews.iloc[0].state
    country = reviews.iloc[0].country
    types = reviews.iloc[0].types
    
    review = {'beer_id': beer_id, 
              'beer_name': beer_name, 
              'beer_style': beer_style, 
              'look': look, 
              'smell': smell, 
              'taste': taste, 
              'feel': feel, 
              'overall': overall,
              'availability': availability,
              'abv': abv,
              'brewery_id': brewery_id,
              'brewery_name': brewery_name,
              'city': city,
              'state': state,
              'country': country,
              'types': types}

    agg = agg.append(review, ignore_index=True)
```
