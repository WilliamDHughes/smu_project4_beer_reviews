```python
import pandas as pd
```

```python
breweries = pd.read_csv("Resources/Clean/texas_breweries.csv")
```

```python
beers = pd.read_csv("Resources/Clean/texas_beers.csv")
```

```python
reviews = pd.read_csv("Resources/Clean/texas_reviews.csv")
```

```python
breweries.info()
```

```python
beers.info()
```

```python
reviews.info()
```

```python
merged1 = pd.merge(reviews, beers, left_on='beer_id', right_on='id')
merged1.info()
```

```python
merged1 = merged1.drop('id', axis=1)
```

```python
merged1.info()
```

```python
merged1.head()
```

```python
merged2 = pd.merge(merged1, breweries, left_on='brewery_id', right_on='id')
merged2.info()
```

```python
merged2.head(20)
```

```python
merged2 = merged2.drop('id', axis=1)
```

```python
merged2.head(20)
```

```python
merged2 = merged2.rename(columns={'name_x': 'beer_name', 'name_y': 'brewery_name', 'style': 'beer_style'})
```

```python
reviews_merged = merged2.reindex(columns=["beer_id", "beer_name", "style", "look", "smell", "taste", "feel", "overall", "availability", "abv", "brewery_id", "brewery_name", "city", "state", "country", "types"])
reviews_merged.head(20)
```

```python
reviews_merged.to_csv('Resources/Clean/reviews_merged.csv', index=False)
```
